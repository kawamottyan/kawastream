```python
import matplotlib.pyplot as plt
```

```python
scores = {'Benchmark': 0.0091, 'GRU4Rec': 0.0219}
```

```python
plt.figure(figsize=(8, 6))
plt.bar(scores.keys(), scores.values(), color=['blue', 'green'])
plt.xlabel('Model')
plt.ylabel('NDCG@100 Score')
plt.title('Comparison of NDCG@100 Scores: Benchmark vs GRU4Rec')
plt.ylim(0, 0.03)

for index, value in enumerate(scores.values()):
    plt.text(index, value + 0.001, f'{value:.4f}', ha='center')

plt.show()
```

```python
benchmark_score = scores['Benchmark']
percentage_scores = {model: (score / benchmark_score) * 100 for model, score in scores.items()}
```

```python
new_y_axis_scale = range(50, 301, 50)

plt.figure(figsize=(8, 6))
plt.bar(percentage_scores.keys(), percentage_scores.values(), color=['blue', 'green'])
plt.xlabel('Model')
plt.ylabel('Relative NDCG@100 Score (%)')
plt.title('Comparison of Relative NDCG@100 Scores: Benchmark vs GRU4Rec')
plt.yticks(new_y_axis_scale)
plt.ylim(50, 300)

for index, value in enumerate(percentage_scores.values()):
    plt.text(index, value + 10, f'{value:.1f}%', ha='center')

plt.show()
```
