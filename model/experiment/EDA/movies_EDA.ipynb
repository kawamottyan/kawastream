```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
from matplotlib.colorbar import ColorbarBase
from matplotlib.cm import cool
import seaborn as sns
from wordcloud import WordCloud
```

```python
df = pd.read_csv('../../data/movies_df.csv')
```

```python
sns.histplot(df['popularity'], kde=True)
plt.title('Distribution of Popularity')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.show()
```

```python
df['is_adult'].value_counts().plot(kind='bar')
plt.title('Distribution of Adult Movies')
plt.xlabel('Is Adult')
plt.ylabel('Count')
plt.show()
```

```python
df['release_date'] = pd.to_datetime(df['release_date'])
df.set_index('release_date', inplace=True)
df.resample('Y').size().plot()
plt.title('Number of Movies Released Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.show()
```

```python
numeric_columns = ['budget', 'revenue', 'popularity', 'runtime', 'vote_average', 'vote_count']
plt.figure(figsize=(10, 8))
sns.heatmap(df[numeric_columns].corr(), annot=True, fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
genre_columns = [col for col in df.columns if col.startswith('genre_')]
for col in genre_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)
genre_counts = df[genre_columns].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 8))
explode = [0.1 if i < 3 else 0 for i in range(len(genre_counts))]
genre_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, explode=explode, textprops={'fontsize': 12})
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), title="Genres")
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
text = ' '.join(df['overview'].dropna())
wordcloud = WordCloud(background_color='white').generate(text)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Movie Overviews')
plt.show()
```

```python
df = df.reset_index(drop=True)
sns.scatterplot(data=df, x='budget', y='revenue')
plt.title('Budget vs Revenue')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.show()
```

```python
df = df.dropna(subset=['budget', 'revenue', 'vote_average'])
fig, ax = plt.subplots(figsize=(10, 6))
scatter = sns.scatterplot(
    data=df, x='budget', y='revenue', hue='vote_average',
    palette='cool', legend=False, ax=ax)
norm = Normalize(vmin=df['vote_average'].min(), vmax=df['vote_average'].max())
sm = plt.cm.ScalarMappable(cmap='cool', norm=norm)
sm.set_array([])
cbar = fig.colorbar(sm, ax=ax)
ax.set_title('Budget vs Revenue by Rating')
ax.set_xlabel('Budget')
ax.set_ylabel('Revenue')
plt.tight_layout()
plt.show()
```
