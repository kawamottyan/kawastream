```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../../data/ratings_df.csv')
```

```python
df.head(5)
```

```python
df.info()
```

```python
df['userId'] = df['userId'].astype(np.int64)
df['tmdbId'] = df['tmdbId'].astype(np.int64)
df['rating'] = df['rating'].astype(np.float64)
df['timestamp'] = df['timestamp'].astype(np.float64)
```

```python
df.info()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['rating'], bins=20, edgecolor='black')
plt.title('Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
user_rating_counts = df['userId'].value_counts()
plt.figure(figsize=(8, 6))
plt.hist(user_rating_counts, bins=30, edgecolor='black')
plt.title('Number of Ratings per User')
plt.xlabel('Ratings Count')
plt.ylabel('Number of Users')
plt.show()
```

```python
movie_rating_counts = df['userId'].value_counts()
top_movies = movie_rating_counts.head(10)
print(f'top number of reviewed user: {top_movies}')
```

```python
movie_rating_counts = df['tmdbId'].value_counts()
plt.figure(figsize=(8, 6))
plt.hist(movie_rating_counts, bins=30, edgecolor='black')
plt.title('Number of Ratings per Movie')
plt.xlabel('Ratings Count')
plt.ylabel('Number of Movies')
plt.show()
```

```python
movie_rating_counts = df['tmdbId'].value_counts()
top_movies = movie_rating_counts.head(10)
print(f'top number of reviewed movie: {top_movies}')
```

```python
df['datetime'] = pd.to_datetime(df['timestamp'], unit='s')
```

```python
df.groupby(df['datetime'].dt.year)['rating'].mean().plot(kind='line', figsize=(10, 6))
plt.title('Average Rating Over Time')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.show()
```

```python
df.groupby(df['datetime'].dt.year)['rating'].count().plot(kind='line', figsize=(10, 6))
plt.title('Number of Ratings Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Ratings')
plt.show()
```

```python
df.groupby(df['datetime'].dt.to_period('M'))['rating'].mean().plot(kind='line', figsize=(10, 6))
plt.title('Average Rating per Month')
plt.xlabel('Month')
plt.ylabel('Average Rating')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x=df['datetime'].dt.year, y='rating', data=df)
plt.title('Rating Distribution by Year')
plt.xlabel('Year')
plt.ylabel('Rating')
plt.show()
```

```python
df.set_index('datetime')['rating'].resample('D').mean().plot(figsize=(12, 6))
plt.title('Daily Average Rating')
plt.xlabel('Date')
plt.ylabel('Average Rating')
plt.show()
```

```python
df.groupby(df['datetime'].dt.dayofweek)['rating'].mean().plot(kind='bar', figsize=(10, 6))
plt.title('Average Rating by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Average Rating')
plt.xticks(ticks=range(7), labels=['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
plt.show()
```
